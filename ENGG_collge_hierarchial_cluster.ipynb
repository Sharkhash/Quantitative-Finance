```python
#create a DENDROGRAM
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
#CHOOSING AVEARGE LIKNAGE METHOD
```

```python
link_method = linkage(enggdata.iloc[:, 1:6], method = 'average')
```

```python
labelList= np.array (enggdata.Engg_College)
```

```python
dend = dendrogram(link_method, labels = labelList)
```

```python
#CUTTING THE DENDROGRAM WITH SUITABLE CLUSTERS
```

```python
dend = dendrogram(link_method, truncate_mode= 'lastp', p=10, labels= labelList)
```

```python
#IMPORTING FCLUSTER MODULE TO CREATE CLUSTERS
```

```python
from scipy.cluster.hierarchy import fcluster
```

```python
# set criterion as maxclust, then create 3 clusters and store the result in another object ' cluster'
clusters= fcluster(link_method, 3, criterion='maxclust')
clusters
```

```python
df['clusters'] = clusters
```

```python
df.head
```

```python
#CLUSTER FREQUENCY
```

```python
df.clusters.value_counts().sort_index()
```
