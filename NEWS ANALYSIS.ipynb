```python
import requests
from datetime import datetime
import time

# SerpApi Google News API key
api_key = 'hiden'  # Replace with your actual API key
query = 'Apple news'

# URL for Google News API via SerpApi
url = f'https://serpapi.com/search?q={query}&tbm=nws&api_key={api_key}'

# Send the GET request
response = requests.get(url)

if response.status_code == 200:
    results = response.json()

    print(f"Fetching news at {datetime.now()}...\n")

    # Loop through the articles and print the latest ones
    for article in results['news_results'][:10]:  # Limit to 10 latest news
        published_date = article.get('published', 'No date available')

        try:
            # If the date is in string format, parse it and convert to desired format
            published_datetime = datetime.strptime(published_date, '%Y-%m-%dT%H:%M:%S%z')
            formatted_date = published_datetime.strftime('%Y-%m-%d %H:%M:%S')
        except ValueError:
            # If the date cannot be parsed, keep it as is
            formatted_date = published_date

        print(f"Title: {article['title']}")
        print(f"Link: {article['link']}")
        print(f"Published on: {formatted_date}")
        print()

else:
    print("Failed to retrieve news")
```
