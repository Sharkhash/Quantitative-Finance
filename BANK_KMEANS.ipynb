```python
# NOW CREATE WSS
```

```python
WSS = []

```

```python
wss = []  # List to store WSS values

for i in range(1, 11):
    KM = KMeans(n_clusters=i, random_state=42)
    KM.fit(scaled_df)
    wss.append(KM.inertia_)  # Append inertia (WSS) to the list

# Plot the Elbow Curve
plt.plot(range(1, 11), wss, marker='o', linestyle='-')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Within-Cluster Sum of Squares (WSS)')
plt.title('Elbow Method for Optimal K')
plt.show()
```

```python
#CLUSTERING PROFILING
```

```python
kmeans = KMeans(n_clusters=4)
data_df['Clus_kmeans4'] = kmeans.fit_predict(data_df.drop(columns=['Bank'], errors='ignore'))
```

```python
# Remove the 'Bank' column before profiling
clust_profile = data_df.drop(['Bank'], axis=1)

# Compute the mean for each cluster
clust_profile = clust_profile.groupby('Clus_kmeans4').mean()

# Count the number of data points in each cluster
clust_profile['freq'] = data_df.Clus_kmeans4.value_counts().sort_index()

# Display the final profiling result
clust_profile
```
